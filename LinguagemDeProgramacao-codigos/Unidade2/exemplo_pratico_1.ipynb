```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("dados.csv")

print(f"Colunas disponíveis: {df.columns}")
print(df.head())

if 'data' not in df.columns:
   raise KeyError("A coluna 'data' não está presente no arquivo csv.")

novo_df = df[['data','col1', 'col2', 'col2']]
novo_df['data'] = pd.to_datetime(novo_df['data'])
novo_df = novo_df.loc[novo_df['data'] <= pd.to_datetime('2020-01-01')]


df = pd.DataFrame({'ano': ['2012', '2012', '2013', '2013', '2014', '2014', '2015', '2015', '2016', '2016', '2017', '2017', '2018', '2018', '2019', '2019', '2020', '2020'],
                   'meses': ['Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro', 'Janeiro', 'Fevereiro'],
                   'vendas': [7000000, 6000000, 500000, 400000, 300000, 200000, 1000000, 700000, 600000, 500000, 400000, 300000, 200000, 100000, 7000000, 6000000, 5000000, 4000000]})

df_grouped = df.groupby(['ano', 'meses']).mean()
df_unstacked = df_grouped.unstack()

df_unstacked.plot(kind='bar', rot=0)
plt.title('Vendas por Ano e Mês')
plt.xlabel('Ano')
plt.ylabel('Vendas (em unidades)')
plt.legend(title='Meses')
plt.tight_layout()

plt.show()
```
